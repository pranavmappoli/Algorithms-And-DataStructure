```python
#Topological sort
from collections import defaultdict
class GraphT:
  def __init__(self,noNodes):
    self.noNodes=noNodes
    self.map=defaultdict(list)
    self.inDegree=[0 for i in range(noNodes)]
  def addEdge(self, u, v):
    self.map[u].append(v)
    self.inDegree[v]=self.inDegree[v]+1


  def __TopologicalHealper(self,node):
    for child in self.map[node]:
      self.inDegree[child]=self.inDegree[child]-1
      if self.inDegree[child]==0:
        print(child)
        self.__TopologicalHealper(child)

  def TopologicalSort(self):
    for i in range(self.noNodes):
      if self.inDegree[i]==0:
        print(i)
        self.__TopologicalHealper(i)

```

```python
from collections import namedtuple, deque
class RottenOrange:
  def __init__(self,matrix):
    self.matrix=matrix
    self.rows=len(matrix)
    self.column=len(matrix[0])
    orange=namedtuple("orange",["x","y","time"])

  def FindRottenEgs(self):
    rottenEgs=[]
    for x in range(self.rows):
      for y in range(self.column):
        if self.matrix[x][y]==2:
          rottenEgs.append(orange(x,y,0))
    return rottenEgs
  def FindAdj(self,x,y,time):
    adjcent = []
    row = self.rows
    col = self.column
    if (x >= 0 and x < row) and (y >= 0 and y < col):
        if y+1 < col and self.matrix[x][y+1] == 1: 
            adjcent.append(orange(x,y+1,time))
            self.matrix[x][y+1] = 2 
        if y-1 >= 0 and self.matrix[x][y-1] == 1:
            adjcent.append(orange(x,y-1,time))
            self.matrix[x][y-1] = 2
        if x-1 >= 0 and self.matrix[x-1][y] == 1: 
            adjcent.append(orange(x-1,y,time))
            self.matrix[x-1][y] = 2
        if x+1 < row and self.matrix[x+1][y] == 1: 
            adjcent.append(orange(x+1,y,time))
            self.matrix[x+1][y] = 2
    return adjcent

  def findTime(self):
    q=self.FindRottenEgs()
    while(q):
      crntItem=q.pop(0)
      adj=self.FindAdj(crntItem.x,crntItem.y,crntItem.time+1)
      q.extend(adj)
    return crntItem.time

rtime=RottenOrange([[2,1,0,2,1],[1,0,1,2,1],[1,0,0,2,1]])
print(rtime.findTime())
```

```python
#########Snake Lader problem##########
from queue import Queue 
class Cell:
  def __init__(self,val,steps):
    self.val=val
    self.steps=steps
def findSnakeSteps(board):
  boardSize=len(board)
  visited={}
  q=Queue()
  q.put(Cell(1,0))
  visited[1]=True
  while(q):
    crntElem=q.get()
    if crntElem.val==boardSize:
      return crntElem.steps
    for i in range(1,7):
      if crntElem.val+i-1<boardSize and crntElem.val+i not in visited:
        if board[crntElem.val+i-1]==-1:
          q.put(Cell(crntElem.val+i,crntElem.steps+1))
        else:
          if board[crntElem.val+i-1] not in visited:
            q.put(Cell(board[crntElem.val+i-1],crntElem.steps+1))


N = 30
moves = [-1] * N 
moves[2] = 21
moves[4] = 7
moves[10] = 25
moves[19] = 28
moves[26] = 0
moves[20] = 8
moves[16] = 3
moves[18] = 6
findSnakeSteps(moves)
        




```

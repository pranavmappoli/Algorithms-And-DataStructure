```python
#Find the longest prefix of set of a string

class TrieNode:
  def __init__(self,char):
    self.char=char
    self.childrens={}
    self.isEnd=False
class Trie:
  def __init__(self):
    self.root=TrieNode("/")
  def insert(self,word):
    node=self.root
    for c in word:
      if c in node.childrens:
        node=node.childrens[c]
      else:
        node.childrens[c]=TrieNode(c)
        node=node.childrens[c]
    node.isEnd=True

#####################################################################Longest prifix###############################################################

  def findLongestPrefix(self):
    count=0
    node=self.root
    while len(node.childrens)==1:
      if node.isEnd==True:
        return count
      count+=1
      for c in node.childrens:
        node=node.childrens[c]
    return count

    ########################################################## WORD SUGGESTIONS ###################################################################


  def __wordSuggestionsHelper(self,node):
    suggestions=[]
    if node.isEnd==True:
      suggestions.append(node.char)
    for crntNode in node.childrens:
      word=self.__wordSuggestionsHelper(node.childrens[crntNode])
      word=list(map(lambda x:node.char+x,word))
      suggestions.extend(word)
    return suggestions

  def wordSuggestions(self,word):
    node=self.root
    for c in word:
      if c in node.childrens:
        node=node.childrens[c]
      else:
        return False
    suggestions=self.__wordSuggestionsHelper(node)
    return suggestions


```

```python
def minheapify(arr,i=1):
  l=2*i
  r=2*i+1
  heapsize=len(arr)-1
  if l<=heapsize and arr[l]<arr[i]:  
    largest=l
  else:
    largest=i
  if r<=heapsize and arr[r]<arr[largest]: #only r<heapsize, not <=
    largest=r
  if largest!=i:      
    arr[i],arr[largest]=arr[largest],arr[i]
    minheapify(arr,largest)
```

```python
arr=[None,5,1,2,3,4]
minheapify(arr,i=1)
arr
```

```python
#assume left and right of root is alredy in heap and root is not.
def maxheapify(arr,i):
  l=2*i
  r=2*i+1
  heapsize=len(arr)-1
  if l<=heapsize and arr[l]>arr[i]:  
    largest=l
  else:
    largest=i
  if r<=heapsize and arr[r]>arr[largest]: #only r<heapsize, not <=
    largest=r
  if largest!=i:      
    arr[i],arr[largest]=arr[largest],arr[i]
    maxheapify(arr,largest)

#Creating heap
def createheap(arr):
  heapsize=len(arr)
  for i in range(heapsize//2,0,-1): #from last internal node->head
    maxheapify(arr,i)

```

```python
class Node:
  def __init__(self,key,value):
    self.key=key
    self.value=value
    self.prev=None
    self.next=None

class LRU:
  def __init__(self,capacity):
    self.capacity=capacity
    self.lrudict={}
    self.len=0
    self.head=Node(-1,-1) # deflt head,for if someone
    self.tail=self.head
  def insert_LRU(self,key,value):
    if key in self.lrudict:
      if self.tail.key !=key:
        node=self.lrudict[key]
        node.prev.next=node.next
        node.next.prev=node.prev
        self.tail.next=node
        node.prev=self.tail
        self.tail=node
        self.tail.next=None
    else:
      node=Node(key,value)
      self.lrudict[key]=node
      self.tail.next=node
      node.prev=self.tail
      self.tail=node
      self.tail.next=None
      self.len+=1
      if self.len>self.capacity:
        delkey=self.head.next.key
        self.lrudict.pop(delkey)
        self.head.next=self.head.next.next

  def get_LRU(self,key):
    if key not in self.lrudict:
      return -1
    else:
      node=self.lrudict[key]
      val=node.value
      if self.tail.key !=key:
        node.prev.next=node.next
        node.next.prev=node.prev
        self.tail.next=node
        node.prev=self.tail
        self.tail=node
        self.tail.next=None
      return val
```

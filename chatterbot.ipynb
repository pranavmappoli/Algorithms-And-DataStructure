```
#chatbot config
from chatterbot import ChatBot
aptivbot = ChatBot('Aptiv',logic_adapters=['chatterbot.logic.BestMatch'])

from chatterbot.trainers import ChatterBotCorpusTrainer
trainernew = ChatterBotCorpusTrainer(aptivbot)
trainernew.train("chatterbot.corpus.english.greetings")
#manual trainer
from chatterbot.trainers import ListTrainer
trainer = ListTrainer(aptivbot)
```

```
trainer.train([
'I need your assistan',
'Please, Provide me with your order id',
'I have a complaint.',
'Please elaborate, your concern',
'How to create a new travel request',
'Login to your Aptiv portal -> select Travel Authorization->fill the form -> click submit',
'how to check my request status',
'On your Aptiv portal, click on Existing Requisition and type your request number',
'How to cancel a new travel request',
'Login to your Aptiv portal -> select Travel Authorization->fill the form with travel cancellation -> click submit'
])
```

```
from flask import Flask, render_template, request
app = Flask(__name__)

@app.route("/")
def home():    
    return render_template("home.html") 
@app.route("/get")
def get_bot_response():
    
    userText = request.args.get('msg')
    #Chatbackup
    f= open("Chatbackup.txt","a+")
    f.write("\n User: %s" % userText)
    outText=str(aptivbot.get_response(userText))
    f.write("\n Bot: %s" %outText)
    return outText

if __name__ == "__main__":    
    app.run()
```

```python
class Node:
  def __init__(self,data):
    self.data=data
    self.next=None
```

```python
#createliked list from a list

def linkedlist():
  head=None
  crnt=None
  mylist=[int(i) for i in input().split()]
  for data in mylist:
    if data==-1:
      break
    node=Node(data)
    if head==None:
      head=node
      crnt=node
    else:
      crnt.next=node
      crnt=node
  return head

```

```python
#createliked list recursive
def makelist():
  data=int(input())
  if data==-1:
    return None
  crntnode=Node(data)
  nextnode=makelist()
  crntnode.next=nextnode
  return crntnode
```

```python
def printlinkedlist(root):
  if root==None:
    return
  print(root.data)
  printlinkedlist(root.next)
```

```python
#findmodnode
def midnode(root):
  slow=root
  fast=root
  while(fast.next!=None and fast.next.next!=None):
    slow=slow.next
    fast=fast.next.next
  return slow
```

**Doubly Linked list**

```python
class dnode:
  def __init__(self,data):
    self.data=data
    self.next=None
    self.prev=None
  
```

```python
def createdd():
  head=None
  crnt=None
  prev=None
  mylist=[int(i) for i in input().split()]
  for nodeval in mylist:
    if nodeval==-1:
      return head
    node=dnode(nodeval)
    if head==None:
      head=node
      crnt=prev=head
    else:
      crnt.next=node
      prev=crnt
      node.prev=prev
      crnt=node
    


```

```python
def printdlforward(root):
  if root==None:
    return
  print(root.data)
  printlinkedlist(root.next)

def printdlbackward(root):
  if root==None:
    return
  print(root.data)
  printdlbackward(root.prev)

def returnlastnode(root):
  while(root.next!=None):
    root=root.next
  return root

  
```

```python
rootdl=createdd()
```

```python
broot=returnlastnode(rootdl)
printdlbackward(broot)
```

```python
# Insertion in doubly LL
def insertdl(root,pos,val):
  if root==None or pos<1:
    return root
  if pos==1:
    newnode=dnode(val)
    newnode.next=root
    newnode.prev=root.prev
    root.prev=newnode
    return newnode
  root.next=insertdl(root.next,pos-1,val)
  return root
```

```python
#Deletion in doubly LL
def deletedl(root,pos):
  if root==None or pos<1:
    return root
  if pos==1:
    root.next.prev=root.prev
    return root.next

  root.next=deletedl(root.next,pos-1)
  return root
```

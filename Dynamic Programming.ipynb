```python
def __matrixMultiplicationHelper(self,arr,L,R):
        if L>=R:
            return 0
        min=float('inf')
        for k in range(L,R):
            left=self.__matrixMultiplicationHelper(arr,L,k)
            right=self.__matrixMultiplicationHelper(arr,k+1,R)
            crnt=arr[L-1]*arr[k]*arr[R]
            total=left+right+crnt
            if total<min:
                min=total
        return min
```

```python
#######################################Bottom Up ############################3

class Solution:
    
  
    def matrixMultiplication(self, N, arr):
        dp=[[-1 for i in range(N)] for _ in range(N)]
        
        for i in range(N):
            dp[i][i]=0
            
        for cl in range(2,N):
            for l in range(1,N-cl+1):
                r=l+cl-1
                dp[l][r]=float('inf')
                for k in range(l,r):
                    tot=arr[l-1]*arr[k]*arr[r]+dp[l][k]+dp[k+1][r]
                    if tot<dp[l][r]:
                        dp[l][r]=tot
        return dp[1][N-1]
```

```python
################################################palindrome partitioning####################################

class Solution:
    def __init__(self):
      self.dp=[]
    def ispalindrome(self,string,l,r):
        while(l<r):
            if string[l]!=string[r]:
                return False
            l+=1
            r-=1
        
        return True
        
    def helper(self,string,l,r):
        if l>=r:
            return 0
        if self.dp[l][r]!=-1:
          return self.dp[l][r]
        if self.ispalindrome(string,l,r):
            return 0
        minval=float('inf')
        for k in range(l,r):
            crntmin=1+self.helper(string,l,k)+self.helper(string,k+1,r)
            minval=min(crntmin,minval)
        self.dp[l][r]=minval
        return minval
            
            
    def palindromicPartition(self, string):
        self.dp=[[-1]*len(string) for _ in range(len(string))]
        return self.helper(string,0,len(string)-1)
s=Solution()
s.palindromicPartition("abcd")
```

```python
#Box Stacking Problem


class cuboid:
    def __init__(self,height,width,length):
        self.height=height
        self.width=max(width,length)
        self.length=min(width,length)
        self.area=width*length

class Solution:
    def maxHeight(self,height, width, length, n):
        cuboidList=[]
        maxHeightDp=[]
        for h,w,l in zip(list(height),width,length):
            cube1=cuboid(h,w,l)
            cube2=cuboid(w,h,l)
            cube3=cuboid(l,w,h)
            cuboidList.append(cube1)
            cuboidList.append(cube2)
            cuboidList.append(cube3)
        
        cuboidList.sort(key=lambda x:x.area)

        for ob in cuboidList:
            maxHeightDp.append(ob.height)
        for j in range(1,len(maxHeightDp)):
            for i in range(j):
                if cuboidList[i].width<cuboidList[j].width and cuboidList[i].length<cuboidList[j].length and maxHeightDp[j]<maxHeightDp[i]+cuboidList[j].height:
                    maxHeightDp[j]=maxHeightDp[i]+cuboidList[j].height
        return maxHeightDp
```

```python
class Box:
     
    # Representation of a box
    def __init__(self, h, w, d):
        self.h = h
        self.w = w
        self.d = d
 
    def __lt__(self, other):
        return self.d * self.w < other.d * other.w
 
def maxStackHeight(arr, n):
 
    # Create an array of all rotations of
    # given boxes. For example, for a box {1, 2, 3},
    # we consider three instances{{1, 2, 3},
    # {2, 1, 3}, {3, 1, 2}}
    rot = [Box(0, 0, 0) for _ in range(3 * n)]
    index = 0
 
    for i in range(n):
 
        # Copy the original box
        rot[index].h = arr[i].h
        rot[index].d = max(arr[i].d, arr[i].w)
        rot[index].w = min(arr[i].d, arr[i].w)
        index += 1
 
        # First rotation of the box
        rot[index].h = arr[i].w
        rot[index].d = max(arr[i].h, arr[i].d)
        rot[index].w = min(arr[i].h, arr[i].d)
        index += 1
 
        # Second rotation of the box
        rot[index].h = arr[i].d
        rot[index].d = max(arr[i].h, arr[i].w)
        rot[index].w = min(arr[i].h, arr[i].w)
        index += 1
 
    # Now the number of boxes is 3n
    n *= 3
 
    # Sort the array 'rot[]' in non-increasing
    # order of base area
    rot.sort(reverse = True)
 
    # Uncomment following two lines to print
    # all rotations
    # for i in range(n):
    #     print(rot[i].h, 'x', rot[i].w, 'x', rot[i].d)
 
    # Initialize msh values for all indexes
    # msh[i] --> Maximum possible Stack Height
    # with box i on top
    msh = [0] * n
 
    # Compute optimized msh values
    # in bottom up manner
    for i in range(1, n):
        for j in range(0, i):
            if (rot[i].w < rot[j].w and
                rot[i].d < rot[j].d):
                if msh[i] < msh[j] + rot[i].h:
                    msh[i] = msh[j] + rot[i].h
 
    maxm = -1
    for i in range(n):
        maxm = max(maxm, msh[i])
 
    return msh
 
# Driver Code
if __name__ == "__main__":
  height = [4,1,4,10]
  width= [6,2,5,12]
  length= [7,3,6,32]
  arr=[]
  for h,w,l in zip(list(height),width,length):
    arr.append(Box(h,w,l))
    n = len(arr)
maxStackHeight(arr, n)
```

```python
# Find Minimum Path

class Solution:
    def __init__(self):
        self.grid=[]
    def helper(self,i,j,m,n):
        if i>=m or j>=n:
            return float('inf')
        if i==m-1 and j==n-1:
            return self.grid[i][j]
        else:
            crntmin=min(self.helper(i+1,j,m,n),self.helper(i,j+1,m,n))
            return self.grid[i][j]+crntmin
=============================Bottom up=================================

class Solution:
    def minPathSum(self, grid: List[List[int]]) -> int:
        m,n=len(grid),len(grid[0])
        for i in range(m):
            for j in range(n):
                if i==0 and j==0:
                    pass
                elif i==0:
                    grid[i][j]=grid[i][j]+grid[i][j-1]
                elif j==0:
                    grid[i][j]=grid[i][j]+grid[i-1][j]
                else:
                    grid[i][j]=grid[i][j]+min(grid[i][j-1],grid[i-1][j])
        return grid[-1][-1]
```

```python
#Given an m x n binary matrix filled with 0's and 1's, find the largest square containing only 1's and return its area.


####################################################Recursion ##############################

class Solution:
    def __init__(self):
        self.matrix=[]
        self.crntmax=0
        self.cache={}
    def helper(self,i,j,m,n):
        key=(i,j)
        if key in self.cache:
            return self.cache[key]
        if i>=m or j>=n:
            return 0
        elif self.matrix[i][j]=="1":
            crnt= 1+min(self.helper(i+1,j,m,n),self.helper(i,j+1,m,n),self.helper(i+1,j+1,m,n))
        else:
            min(self.helper(i+1,j,m,n),self.helper(i,j+1,m,n),self.helper(i+1,j+1,m,n))
            crnt=0
        if crnt>self.crntmax:
            self.crntmax=crnt
        self.cache[key]=crnt
        return crnt
        
        
    def maximalSquare(self, matrix: List[List[str]]) -> int:
        self.matrix=matrix
        m,n=len(matrix),len(matrix[0])
        self.helper(0,0,m,n)
        area= self.crntmax*self.crntmax
        return area

=================================================================Bottom Up===================================================

class Solution:
    def maximalSquare(self, matrix: List[List[str]]) -> int:
        m,n=len(matrix),len(matrix[0])
        dp=[[0 for i in range(n+1)] for j in range(m+1)]
        crntmax=0
        for i in range(1,m+1):
            for j in range(1,n+1):
                if matrix[i-1][j-1]=="1":
                    dp[i][j]=1+min(dp[i-1][j],dp[i][j-1],dp[i-1][j-1])
                    crntmax=max(crntmax,dp[i][j])
                else:
                    dp[i][j]=0
        return crntmax*crntmax
```

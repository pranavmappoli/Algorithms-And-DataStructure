```python
def ratmaze(x,y,inarray,outarray):
  n=len(inarray)
  if x==y==n-1:
    outarray[x][y]=1
    print(outarray)
    return
  if x<0 or y<0 or x>=n or y>=n or inarray[x][y]==0 or outarray[x][y]==1:
    return
  outarray[x][y]=1
  ratmaze(x+1,y,inarray,outarray)
  ratmaze(x,y+1,inarray,outarray)
  ratmaze(x-1,y,inarray,outarray)
  ratmaze(x,y-1,inarray,outarray)
  outarray[x][y]=0
  return

n=int(input())
inarray=[]
for i in range(n):
  elem=[int(ele) for ele in input().split()]
  inarray.append(elem)
outarray=[[0 for _ in range(n)] for i in range(n)]
ratmaze(0,0,inarray,outarray) 

```

```python
def isplaceble(board,row,col):
  n=len(board)
  i=row
  j=col
  while(i>=0):
    if board[i][j]==1:
      return False
    i-=1
  i=row
  j=col
  while(j>=0):
    if board[i][j]==1:
      return False
    j-=1
  i=row
  j=col
  while(i>=0 and j>=0):
    if board[i][j]==1:
      return False
    i-=1
    j-=1
  i=row
  j=col
  while(i>=0 and j<n):
    if board[i][j]==1:
      return False
    i-=1
    j+=1
    
  return True
  
    
```

```python
def nqueen(board,row,col):
  n=len(board)
  if row>=n:
    print(board)
    return
  
  for col in range(n):
    if isplaceble(board,row,col):
      board[row][col]=1
      nqueen(board,row+1,col)
      board[row][col]=0
  return

n=int(input())
board=[[0 for _ in range(n)] for i in range(n)]
nqueen(board,0,0)
```

```python
def paranthesis(n,output,open,close):
  if open==close==n:
    print(output)
    return
  if open<n:
    output.append('{')
    paranthesis(n,output,open+1,close)
    output.pop()
  if open>close:
    output.append('}')
    paranthesis(n,output,open,close+1)
    output.pop()
  return
paranthesis(3,[],0,0)
```

```python
def permute(str,start,end):
  if start==end:
    print(str)
    return
  for i in range(start,len(str)):
    str[start],str[i]=str[i],str[start]
    permute(str,start+1,end)
    str[start],str[i]=str[i],str[start]
    
# Combinations

def combi(str,out,k,i):
  if len(out)==k:
    print(out)
    return
  if len(str)==i:
    return
  out.append(str[i])
  combi(str,out,k,i+1)
  out.pop()
  combi(str,out,k,i+1)
```

```python
#power set

def powerset(str):
  if len(str)==0:
    return [""]
  crntelem=str[0]
  newelems=powerset(str[1:])
  newset=[]
  for i in newelems:
    newset.append(i)
    newset.append(crntelem+i)
  return newset

powerset("abc")

```

```python
#knight walk problem
def kingnight(Cx,Cy,Dx,Dy,outboard,steps):
  if Cx==Dx and Cy==Dy:
    outboard[Cx][Cy]=1
    return steps
  if Cx<0 or Cy<0 or Cx>=len(outboard) or Cy>=len(outboard) or outboard[Cx][Cy]==1:
    return float('inf')
  outboard[Cx][Cy]=1
  min1=kingnight(Cx+2,Cy-1,Dx,Dy,outboard,steps+1)
  min2=kingnight(Cx+2,Cy+1,Dx,Dy,outboard,steps+1)
  min3=kingnight(Cx-2,Cy+1,Dx,Dy,outboard,steps+1)
  min4=kingnight(Cx-2,Cy-1,Dx,Dy,outboard,steps+1)
  min5=kingnight(Cx+1,Cy+2,Dx,Dy,outboard,steps+1)
  min6=kingnight(Cx+1,Cy-2,Dx,Dy,outboard,steps+1)
  min7=kingnight(Cx-1,Cy+2,Dx,Dy,outboard,steps+1)
  min8=kingnight(Cx-1,Cy-2,Dx,Dy,outboard,steps+1)
  crntmin=min(min1,min2,min3,min4,min5,min6,min7,min8)
  outboard[Cx][Cy]=0 #backtracking
  return crntmin

n=4
board=[[0 for _ in range(n)] for i in range(n)]
kingnight(1,1,0,0,board,0)

  
```

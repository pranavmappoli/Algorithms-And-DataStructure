```python
class Node:
  def __init__(self,data):
    self.data=data
    self.left=None
    self.right=None
```

```python
#Creating BST Tree

def BST(root,data):
  if root==None:
    root=Node(data)
    return root
  if data<root.data:
    root.left=BST(root.left,data)
  else:
    root.right=BST(root.right,data)
  return root

root=None
l=[10,20,40,70,80]
for i in l:
  root=BST(root,i)
  
```

```python
def search(root,val):
  if root==None:
    return False
  if root.data==val:
    return True
  if val>root.data:
    return search(root.right,val)
  else:
    return search(root.left,val)


```

```python
def checkbst(root):
  if root==None:
    return True,0,float('inf')
  flagl,maxl,minl=checkbst(root.left)
  flagr,maxr,minr=checkbst(root.right)
  crntflag=False
  if flagl==flagr==True:
    if maxl>=root.data or minr<root.data:
      crntflag= False
    else:
      crntflag= True
  else:
    crntflag= False
  maxl=max(root.data,maxr,maxl)
  minr=min(root.data,minl,minr)
  return crntflag,maxl,minr
  
  
```

```python
def checkbst2(root,left=float('-inf'),right=float('inf')):
  if root==None:
    return True
  crntnode=False
  if root.data<=right and root.data>left:
    crntnode=True

  lefttree=checkbst2(root.left,left,root.data)
  righttree=checkbst2(root.right,root.data,right)
  if lefttree==crntnode==righttree==True:
    return True
  else:
    return False

```

```python
def tracepath(root,k):
  if root==None:
    return -1
  if root.data==k:
    return [k]
  left=tracepath(root.left,k)
  right=tracepath(root.right,k)
  if left!=-1:
    left.append(root.data)
    return left
  if right!=-1:
    right.append(root.data)
    return right
  return -1
```

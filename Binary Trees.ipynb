```python
#diameter

def getBTDiameter(root):
    if root==None:
        return 0,0
    left,lefth=getBTDiameter(root.left)
    right,righth=getBTDiameter(root.right)
    height=1+max(lefth,righth)
    crntdiameter=lefth+righth+1
    return max(crntdiameter,left,right),height
```

```python
#######LEVEL INPUT


from queue import Queue
def levelinput():
  q=Queue()
  print("Enter Root")
  rootval=int(input())
  root=Node(rootval)
  q.put(root)
  while(not q.empty()):
    qrntnode=q.get()
    print("Enter left value of :",qrntnode.data)
    leftval=int(input())
    if leftval!=-1:
      leftnode=Node(leftval)
      qrntnode.left=leftnode
      q.put(leftnode)

    print("Enter right value of :",qrntnode.data)
    rightval=int(input())
    if rightval!=-1:
      rightnode=Node(rightval)
      qrntnode.right=rightnode
      q.put(rightnode)
  return root
  
```

```python
def leveloutput(root):
  q=Queue()
  if root!=None:
    q.put(root)
  else:
     return -1
  while(not q.empty()):
    crntval=q.get()
    if crntval!=None:
      print(crntval.data,end=" ")
      q.put(crntval.left)
      q.put(crntval.right)
```

```python
def leef(root):
  if root==None:
    return
  if root.left.left==root.left.right==None:
    root.left=None
  if root.right.left==root.right.right==None:
    root.right=None
  leef(root.left)
  leef(root.right)
```

```python
def mirror(root):
  if root==None:
    return
  root.left,root.right=root.right,root.left
  mirror(root.left)
  mirror(root.right)
  return
mirror(root)
```

```python
def balanced(root):
  if root==None:
    return 0,0
  h1,lh1h2=balanced(root.left)
  h2,rh1h2=balanced(root.right)
  h=1+max(h1,h2)
  return h,max(h1+h2,lh1h2,rh1h2)
  
```

```python
def rootfinder(preord,inord):
  j=0
  while(j<len(preord)):
    for i in range(len(inord)):
      if preord[j]==inord[i]:
        return inord[i],i
    j+=1
  return -1,-1
```

```python
def maketree(preord,inord):
  if len(inord)<1:
    return None
  rootval,rootvalpos=rootfinder(preord,inord)
  print(rootval,rootvalpos)
  root=Node(rootval)
  lefttree=maketree(preord,inord[:rootvalpos])
  righttree=maketree(preord,inord[rootvalpos+1:])
  root.left=lefttree
  root.right=righttree
  return root

```

```python
def sumroot(root,k):
  print("hai")
  if root==None:
    return
  data=[root.data]
  if k==0:
    return data
  data.extend(sumroot(root.left,k-root.data))
  data.extend(sumroot(root.right,k-root.data))
```
